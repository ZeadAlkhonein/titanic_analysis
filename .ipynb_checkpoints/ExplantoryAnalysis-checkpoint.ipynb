```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
new_df = pd.read_csv('Clean_dataframe.csv')
age_group = pd.read_csv('age_group.csv')
group_class = pd.read_csv('group_class.csv')
```

```python
plt.title('Group by age survivle rate')
sns.barplot(data=age_group,x='age_stage',y='Survived',order=['Young','Young Adult','Adult','senior']);
```

## what we see here that more young people have more probability to survive than older people but we also see young adult die more than adult and the least survival rate is senior people

```python
plt.title('Fare based on class')
sns.barplot(data=new_df,x='Pclass',y='Fare');
```

```python
sns.barplot(data=group_class,x='Pclass',y='Survived');
```

### People with higher classes have more survival chance to live
